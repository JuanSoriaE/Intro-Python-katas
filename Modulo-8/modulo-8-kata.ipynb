```python
planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1
}
```

```python
moons = planet_moons.values()
planets = len(planet_moons)
moonsCounter = 0

for value in moons:
    moonsCounter = moonsCounter + value

averageMoons = moonsCounter / planets
print(round(averageMoons, 1))
```
