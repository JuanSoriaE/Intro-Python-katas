```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']

print(len(planets))
```

```python
planets.append('Pluto')

print(len(planets))
print(planets[-1])
```

```python
planet = input("Type the planet name you're interested: ")

index = planets.index(planet)
print('Nearest planets to the Sun:')
print(planets[0:index])
print('\nFarest planets to the Sun:')
print(planets[index + 1:])
```
