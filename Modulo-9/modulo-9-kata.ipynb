```python
def report(prelaunchTime, flightTime, destination, externalTank, internalTank):
    return f'Destination: {destination}\nTotal travel time: {prelaunchTime + flightTime}\nTotal fuel: {externalTank + internalTank}'
```

```python
print(report(10, 120, 'Moon', 100, 200))
```

```python
def report(destination, *minutes, **fuelReservoirs):
    return f'Destination: {destination}\nTotal travel time: {sum(minutes)}\nTotal fuel: {sum(fuelReservoirs.values())}'
```

```python
print(report('Moon', 10, 120, externalTank=100, internalTank=200))
```

```python
def report(destination, *minutes, **fuelReservoirs):
    finalReport = f'Destination: {destination}\nTotal travel time: {sum(minutes)}\nTotal fuel: {sum(fuelReservoirs.values())}\n'

    for fuelTank, capacity in fuelReservoirs.items():
        finalReport += f'{fuelTank} tank > {capacity} left.\n'
    
    return finalReport
```

```python
print(report('Moon', 10, 120, externalTank=100, internalTank=200))
```
